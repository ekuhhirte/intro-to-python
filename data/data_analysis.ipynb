```python
import faker

fake = faker.Faker()

print(fake.email())
print(fake.address())
print(fake.company())
```

```python
# Create a user
import json #imports json so we can use the json dump

users = []  # list of users

for i in range(1000): #skips the inidices and just runs it a thousand times
    user = {
        "username": fake.unique.user_name(),
        "name": fake.name(),
        "email": fake.email(),
        "address": fake.address(),
        "age": fake.random_int(min =18, max = 90)
    }
    users.append(user)

with open("users.json", "w") as file: #creates a file if it doesn't exist, if it does exist it overwrites it
    json.dump(users, file, indent=4) #indentation is set so it looks a bit better

#print(users)
```

```python
# Create an ad campaign
# Start and end dates
# Target age demographic
# Currency
import random
from datetime import date, timedelta

def get_start_end_dates():
    duration = random.randint(1, 2*365)
    offset = random.randint(-365, 365)
    start = date.today() - timedelta(days=offset)
    end = start + timedelta(days=duration)

    return start.strftime("%Y %m %d "), end.strftime(" %Y %m %d ")

#print(get_start_end_dates())

# Create target age demographic
def get_age_range():
    age = random.randrange(20, 46, 5) #min 20, max 46, step 5
    difference = random.randrange(5, 21, 5) #min 5, max 20, step 5

    return f"{age}-{age + difference}"

#print(get_age_range())

def get_currency():
    return random.choice(("GBP", "USD", "EUR")) #choice picks a random out of the passed iteratable, in this case a tuple

def get_campaign_name():
    start, end = get_start_end_dates()
    return f"{start}_{end}_{get_age_range()}_{get_currency()}"
    # <start date>_<end date>_<target age>_<currency>

print(get_campaign_name())
```

```python
def get_campaign_data():
    name = get_campaign_name()
    budget = random.randint(10**3, 10**6)
    spent = random.randint(10**2, budget)

    return {
        "campaignName": name,
        "campaignBudget": budget,
        "campaignSpent": spent
    }

print(get_campaign_data())
raw_data = []

def get_raw_data(users):
    global raw_data
    raw_data = []
    campaigns = []
    for user in users:
        """for i in range(random.randint(5,10)):
            #campaigns = [get_campaign_data() for _ in range(random.randint(2,8))] # _ is used to indicate a variable that is just being thrown away
           campaign_data = get_campaign_data()
            age_range = campaign_data["campaignName"].split("_")[2].split("-") # converts the campaign name to a list of the individual data, 
            #then grabs the 2nd one which should be the age range. Then it splits that  and grabs the start which is the minimum age
            min_age = int(age_range[0])
            max_age = int(age_range[1])

            if user["age"] >= min_age and user["age"] <= max_age:
                campaigns.append(campaign_data)

        #raw_data.append({"user:" user, "campaigns": campaigns})"""
    campaigns = [get_campaign_data() for _ in range(random.randint(2,8))]
    raw_data.append({"user": user, "campaigns": campaigns})
        
    return raw_data

with open("rawdata.json", "w") as file:
    json.dump(get_raw_data(users), file, indent=4)

campaign_data = []
for data in raw_data:
    for campaign in data["campaigns"]:
        campaign["user"] =  data["user"]
        campaign_data.append(campaign)

with open("campaign_data.json", "w") as file:
    json.dump(campaign_data, file, indent=4)

```
